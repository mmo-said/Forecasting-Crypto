# 2) Regression and Classification problems

##### Supervised Learning

Supervised learning is a form machine learning where an algorithm instructs the learning system using clearly defined inputs, which can be measured. These will have an influence over an output. The goal of supervised learning is to provide the best prediction values of outputs using the input data that have been fed into the learning system [4]. 

There are two main types of supervised learning problems. Classification and Regression. 

#### Classification

Classification objective is to predict discrete output variables y a set of x inputs to predict. Where
\begin{equation}
    y \epsilon [1,...C]
\end{equation}

 C is defined as the number of classes. If C=2 then this is called a binary classification. If C is bigger than 2 then it is considered a multi class classification [5]. If however on the other hand the class labels are not mutually exclusive, this will be defined as a multi label classification. Murphy (2016)[5] defined this type of classification as predicting multiple related binary class label. An example of a multi-class classification can be to classify pictures of animals into their species. Or in case of financial markets, the classification of a buy or sell signal depending on the current price action.

 
 #### Estimating F
 
 One method to formalize what type of class problem we are facing is through Function Approximation. We assume y = f(x) for some unknown function f and the goal of the learning is to estimate the function f given a labeled training set in order to make predictions using \begin{equation}
       \hat{y} =\hat{f}(x)
 \end{equation}
 
 where $\hat{f}$ represents our estimate for f, $\hat{y}$ represents the output prediction for y. The accuracy of $\hat{y}$ for predicting y depends on two quantities, which are called reducible error and irreducible error [6]. Generally, $\hat{f}$ will not be a perfect estimate f and this inaccuracy will produce some error. The first type of error can be reduced since it is a reducible error which in turn will improve the accuracy of $\hat{f}$. However the second type cannot, this is due Y being a function of $\epsilon$ which by definition cannot be predicted by using x [6]. Hence we have an irreducible error. This can be shown as follows: 

\begin{equation}
    E(Y - \hat{Y})^2 = [f(X) - \hat{f}(x)]^2 - Var(\epsilon)
\end{equation}
 
 Where  $E(Y - \hat{Y})^2$ represents the average expected value of the squared difference between the predicted and actual value of Y. $\Var(\epsilon)$ is represents the variance associated with the error term which is the irreducible error and the $\[f(X) - \hat{f}(x)]^2$ represents the reducible error. Having a low reducible error is vital especially when predicting price returns for buy or sell signals in the financial markets. Errors that stem from buying or selling signals can be extremely costly hence it will be important to have a low reducible error in order to minimise costly mistakes. 
 
 
#### Regression
Regression problems are just like classification problems, the only difference is that the response variable is continuous. Some examples of real world regression problems include, predicting next day stock price, predicting age of a viewing on a given program, or predicting the temperature of a given area. These problems can be linear or non linear, this research will be focused on solving linear regression problems. 

Linear regression problems are used to find a linear relationship between independent and dependent variables on a given data-set. It is used to measure the predicted effects and model them against input variables to then evaluate the model in order to establish a linear relationships between variables that are dependent and independent [12], which is why it may serve to be useful when attempting to predict price of a stock given a set of continuous variables such as volume, volatility, etc...  The problem we will be tackling for the most part of this research is a regression problem, however before we delve into them, I will briefly explain KNN and how it can be used to solve classification problems. 
 
##### K Nearest Neighbor (KNN)} 
 
 In simple terms, the algorithm looks at the nearest neighbor of K instance being classified and predict the next classification based on that.
This classification method is a non parametric method that bases its prediction on the target output of the K nearest neighbor of the given query point (5). Specifically given a certain data point, we compute the Euclidean distance between the point and all points in the training set. We then pick the closest K training point and set the prediction as the average of the target output values for these K points. 

\begin{equation}
    \hat{y} (x) = 1/k \sum yi
\end{equation}

where $\hat{y}(x)$ is the neighborhood of x defined by the k closest points in xi in the training sample. Closeness will imply a metric, which will be referred to the Euclidean distance which is essentially the squared distances between two points [7]. The algorithm essentially finds the Kth observation with the xi closest to x in input space and the average of their distance. Based on the response the model then predicts $\hat{y}$  based on the closest k observation given by the Euclidean distance. The advantage of using this method is do not rely on any stringent assumptions about the underlying data and can usually adapt to any situation. However the drawback with that as K grows the method becomes less flexible and produces decision boundaries that is close to linear which can correspond to a low variance and high bias in classification. 


 
#### Model Training Prediction
 
 In our example, which I will detail more below, we attempt to predict whether the price of S&P 500 will go up or down, based on the variables; price of 5 lag operators, volume and today price.  A way to handle this problem is to return a probability distribution over the possible binary class label up or down. This probabilistic output will allow to compute our best guess to the true label. 
 
 \begin{equation}
     \hat{y} = \hat{f}(x) = \arg\max_{C} \sum p(y=C\midx,D)
 \end{equation}
 
 where D represents the training set and C being the class label. This notation makes explicit that the prediction y which represents either a price going up or down is a conditional probability of the argmax sum of the class labels conditional on x variables and the training set D.  This corresponds to the most probable class label, known as the mode of the distribution[5]. 
 
Returning a probabilistic allows to handle ambiguous cases, in the example of classifier whether to enter or exit a trade given a time period, we would like to have an accurate reading, if we encounter $P(\hat{y}| x, D)$ the features we are using. 
 
 
Another method utilised to determine the label set for the test instance is through maximum a posterior (MAP) principle [8]. 

When applying KNN, the challenge will be to choose an appropriate value for K and the success of the classification is very much dependent on this value [9]. One way to deal with this challenge in order for KNN to be less dependent on the choice of k, Wang [10] proposes to analyse multiple sets of nearest neighbors rather than just one set of k-nearest neighbors. This will be based on a contextual probability $\hat{y}$ and the method is to aggregate the support of multiple sets of nearest neighbors for various classes to give a more reliable support value. This in essence reveals a better class of x.

# 3. Random Forest

Random Forrest classification is a well known machine learning method for developing prediction models, often the goal is of a prediction model is to reduce the number of variables needed to obtain a prediction in order to reduce the burden of data collection and improve efficiency. However the unique characteristic of random forests is that they are a combination of tree predictors in such way that each tree depends on the value of a random vector sampled independently and with all the same distribution for all trees in the forest. Hence it is computationally efficient and can operate quickly over large data sets [16]. 

The algorithm was proposed by L.Breiman in 2001 with an aim to take advantage of the large and sheer size of modern data sets that can learn with the scale and volume of information available while still maintaining sufficient statistical efficiency [17]. 

To give a formal definition let us start by assuming a training set T with A attributes, n instances and Tk as a bootstrap training set sampled from T with replacement and containing m random attributes such that (m=< A) with n instances [16]. 

In such scenario, a random tree is a tree drawn at random from a set of possible trees with m random attributed at each node. The element of randomness means that each tree has an equal chance of being sampled this means that each random tree can be efficiently generated and the combination of large sets of random trees may generally lead to accurate models [16][18]. 

Hence a Random Forest is formally defined as follows [19]. It is a classifier consisting of a collection of tree structured classifiers {hk(x, Tk)}, where k = 1, 2, ...., L and Tk are independent identically distributed random samples with each tree casts a unit vote for the most popular class at input x.

As mentioned earlier, Another feature of Random Forests is that they employ the same method bagging to produce random samples of training sets for each random tree. Each new training set is built using the new subset and a random attribute selection. The best split on the random attributes selected is used to split the node. There are two reasons to use bagging method [16]. One use seems to enhance performance when random attributes are used and second reason is bagging can be used to give ongoing estimates of the generalization error of the combined ensemble of trees as well as estimates for the strength and correlation. The error of a forest depends on the strength of the individual trees in the forest and the correlation between any two trees in the forest. The strength can be interpreted as a measure of performance for each tree. Increasing the correlation increases the forest error rate while increasing the strength of the individual tree decreases the forest error rate, as such reducing the number of random attributes selected reduces both the correlation and the strength [16]. 


#### What determines the optimal number of trees in a Random Forest?

Oshiro et al. 2018 analysed 29 data sets and examined the performance of random forests algorithms using 256, 512, 1024, 2048, 4096 trees. They found that the mean and median of the AUC (Area Under the ROC Curve (AUC) provides an aggregate measure of performance across all possible classification thresholds), which showed that there presents no major changes from 64 trees. Based upon their experiment, a range between 64 and 128 trees in a forest obtained a good balance between AUC, processing time and memory usage [16].

# 4. Gradient Boosting Method (GBM)

Gradient Boosting Method is an ensemble approach that relies on combining a large number of relatively weak simple models to obtain a stronger ensemble prediction [20]. Unlike random forest which rely on simple averaging of tree models in the ensemble. The boosting method is based on a different strategy of ensemble formation. The premise is to add new models to the ensemble sequentially with each iteration, a new weak base-learner model is trained with respect to the error of the whole ensemble learnt so far [20]. The algorithm was developed by Jerome H. Friedman [21] with an aim to provide a more accurate estimate of the response variable, the principle was to construct new base-learners to be maximally correlated with the negative gradient of the loss function that is associated with the whole ensemble [20][22]. Each loss function can be arbitrary according to the particular task that you want the algorithm to learn. The choice of the loss function is often influenced by the demands of specific characteristics of the conditional distribution. To use an arbitrary loss function, one has to specify both the loss function and the function to calculate the corresponding negative gradient. These two functions can be directly applied into the GBM algorithm [21]. This high flexibility makes the GBMs highly customizable to any particular data driven task as well as it adds a lot of freedom into the model design thus making the choice of the most appropriate loss function a matter of trial and error. This feature also provides a higher accuracy results compared to conventional single strong machine learning model. 



#### Gradient Boosting Algorithm

This section will provide the methodology behind gradient boosting algorithms as derived by Friedman [21]. 

Given a training data-set $D=\{\mathbf{x}_i,y_i\}_{1}^N$, where the goal of the gradient boosting is to find an approximation, $hat{F}(\mathbf{x})$ of function $F^(\mathbf{x})$ which maps instances x to their output y by minimising the expected calue of a given loss function $L(y, F(\mathbf{x}))$. 

Gradient boosting builds an additive approximation of $F^(\mathbf{x})$ as a weighted sum of fucntion such that: 

\begin{equation}
    F_m(\mathbf{x}) = F_{m-1}(\mathbf{x}) + \rho _m h_m(\mathbf{x})
\end{equation} 



where pm is the weight of the mth function, $h_m(\mathbf{x})$. These functions are the models of the ensemble decision trees. The approximation is constructed iteratively where $F^(\mathbf{x})$ is such:


\begin{equation}
    F_{0} ({\mathbf{x}}) = \arg \min _{\alpha } \sum\limits_{{i = 1}}^{N} L (y_{i} ,\alpha ).
\end{equation}

And the models that are expected to minimise are: 

\begin{equation}
    (\rho _{m} ,h_{m} ({\mathbf{x}})) = \arg \min _{{\rho ,h}} \sum\limits_{{i = 1}}^{N} L (y_{i} ,F_{{m - 1}} ({\mathbf{x}}_{i} ) + \rho h({\mathbf{x}}_{i} )) 
\end{equation}

Each $h_{m}$ can be seen as a greedy step in a gradient descent optimization for $\hat{F}$. 
Hence each model $h_{m}$ is trained on a new data-set $D=\{\mathbf{x}_i,r_{mi}\}_{i=1}^N$. And the residuals is calculated by 

\begin{equation}
    r_{mi} = \left[ \frac{\partial L(y_i,F(\mathbf{x}))}{\partial F(\mathbf{x})} \right] _{F(\mathbf{x})=F_{m-1}(\mathbf{x})}
\end{equation} 

The value of pm is then computed by solving for a line search optimization problem. This algorithm can suffer from over-fitting if the iterative process is not properly regularized [21][23]. For the loss functions if the model $h_{m}$ fits the pseudo-residuals perfectly, then the next iteration of the pseudo residuals become zero and the process terminates. In order to control the additive process of the gradient boosting, several regularization hyper-parameters are considered: 

- Shrinkage $\nu$ also known as the (learning rate), which reduces each gradient descent step $F_m(\mathbf{x}) = F_{m-1}(\mathbf{x}) + \nu \rho _m h_m(\mathbf{x})$ with $\nu$ is usually set to 0.1.

- Maximum depth of the tree(max depth): this parameter limits the complexity of the trained model. In the case of the decision trees, one can limit the depth of the trees or the minimum number of instances necessary to split a node. The depth is generally limited to 5-10. 

- Sub sampling rate (subsample): This parameter randomises the base learners which can improve the generalization of the ensemble. Contrary to random forest this is usually carried out without replacement. 

- Number of features (max features): similar to random forest this considers the number of features when looking for the best split. 

-Minimum samples (min samples split): This parameter considers the minimum number of samples required to split an internal node. 

##### Overview of the types of Gradient Boosting Algorithms
In this section I will discuss the family of gradient boosting algorithms and some of the essentially hyper parameters that characterises each algorithm. 

##### XGBoost 
XGBoost is a decision tree ensemble based on gradient boosting designed to be highly scalable. Similarly to gradient boosting, XGBoost builds an additive expansion of the objective function by minimizing a loss function [23]. XGBoost is focused only on decision tree as base classifiers, a variation of the loss is used to control the complexity of the trees. 

\begin{equation}
    L_{xgb}= & {} \sum _{i=1}^N L(y_i, F(\mathbf{x}_i)) + \sum _{m=1}^M \varOmega (h_m)
\end{equation} 

\begin{equation}
    \varOmega (h)=\, & {} \gamma T + \frac{1}{2} \lambda \Vert w\Vert ^2 
\end{equation}

where T is the number of leaves of the tree and w are the output of the leaves. This loss function can be integrated into the split criterion of decision trees leading to a pre-pruning strategy. higher values of $\gamma$ results in simpler trees [23].  The value of the $\gamma$ controls the minimum loss reduction gain needed to split an internal node.

XGBoost models are trained quite fast and require less storage space. The reason for that is because of the regularization hyper parameter feature which is shrinkage, this reduces the steps size in additive expansion hence the complexity of the trees can be limited using other strategies as the depth of trees. Another technique that helps reduce training speed and over-fitting is randomization which include random sub-samples to train individual trees and column sub-sampling at tree and tree node levels. Lastly, XGBoost can be extended to any user defined loss function by defining a function that outputs the gradient and the hessian (second order gradient) and passing it through the "objective" hyper-parameter** [23]. 

Furthermore, XGBoost has a fast training speed and can training and test a dataset and provide a prediction within seconds. The reason for that is because XGBoost implements several methods to increment the training speed of decision trees not directly related to ensemble accuracy. The algorithm focuses on reducing the computational complexity for finding the best split which is the most time consuming section of decision tree construction algorithm [23][24].



##### LightGBM 
LightGBM is an extensive library that implements gradient boosting with some variants, the implementation has been focused on creating a computationally efficient algorithm which is based on the pre-computation of histogram features as in XGBoost [25]. The library also includes dozens of hyper-parameters that allow the model to work in a  wide variety of cases and scenarios. The implementation can work both in GPU and CPU as well as basic gradient booting and many types of randomization[23].

The main feature that sets it apart from other boosting algorithms is that it mainly uses Gradient-based-One-Side-Sampling (GOSS) and Exclusive Feature Bundling (EFB). GOSS is a subsampling technique used to create the training sets for building the base trees in the ensemble [25]. One technique called AdaBoost aims at incrementing the importance of those instances with higher uncertainty in thier classifications, which identifies those instances with higher gradient. When the GOSS option is implemented, the training sets for the base learners are composed of the top franction of the instances with the largest gradient (a) plus a random sample fraction (b) retrieved from the instances with the lowest gradients. On the other hand EFB technique bundles sparse features into a single feature without losing any information when those features do not have a non zero value simultaneously[23]. Both of those techniques provide training speed gains and for the majority part, it is why LightGBM algorithm is known for its speed and efficiency. 


#### CatBoost
CatBoost stands for Categorical boosting similar to other boosting algorithms it builds a predictor model by iterative combining of decision trees as base predictors via a greedy procedure that corresponds to a gradient descent in a function space [27]. However what separates CatBoost from other algorithms is that it can handle categorical features and takes advantage of dealing with them during training. The way this is implemented is by having the categorical features substituted by a numeric feature that measures the expected target value for each category[23]. In order to avoid over fitting it implements ordered boosting which is a permutation driven alternative to the classic boosting algorithm. On small data sets, the gradient boosting can quickly over-fit however in CatBoost there are special built in metrics and modifications that that handles the problem of over-fitting in small data sets much more effectively than other boosting algorithms[27][23]. 

The way it implements that is by reducing the prediction shift that occurs during training. The distribution shift is the depart of $F(\mathbf{x_i})|\mathbf{x_i}$ with xi being the training instance. The shift occurs because during training, gradient boosting is using the same instances for the estimation of both the gradient and the models that minimize those gradients. The solution proposed in CatBoost [26] is to estimate the gradients using a sequence of base models that do not include the instance in their training set. This is done by the introduction of a random permutation in the training instance. The principle in CatBoost is to build i = 1, ....., N base tree models per each of M boosting iteration. The ith model of the mth iteration is training on the first i instances of the permutation and is used to estimate the gradient of the i+1 instance for the (m+1)th boosting iteration. This is so the initial random permutation of this prices is random and is repeated using a different random permutation per iteration. Hence CatBoost optimises the predictive model through trees that are grown by extending level wise all leaf nodes using the same split condition. 

## Implementation of GBM (XGBoost)

For the implementation section of XGBoost and the other algorithms, we will use the the same dataset as used in random forest however the only difference is that we will add new features, since we are predicting a different label now (Target) it may prove useful to add in new features in order to test whether the new features can improve prediction without the risk of overfitting.

The features that we utilised were as follows: 

- Upper shadow: This is the difference betwen the highest minute price and the lowest price.  

- Lower shadow: this is the price difference between the lowest minute price and the highest price.

- Open to Close: Close price devided by Open price. Measure the price of the trading day from open to close

- High to Low: High price devided by low price.

- High to mean: difference between the high price of the day and mean price. 

- Volume to count: Volume devided by the next minute count. 

```python
# feature engineering and setting the features

def get_features(eth_data):
  eth_data['upper_shadow'] = eth_data['High'] / eth_data[['Close', 'Open']].max(axis=1) #difference between highest price and lowest price 
  eth_data['lower_shadow'] = eth_data[['Close', 'Open']].min(axis=1) / eth_data['Low'] #price difference between the lowest price and highest price
  eth_data['opentoclose'] = eth_data['Close'] / eth_data['Open'] #difference between open and close for each day 
  eth_data['high2low'] = eth_data['High'] / eth_data['Low'] # difference between high to low close price 
  mean_price = eth_data[['Open', 'High', 'Low', 'Close']].mean(axis=1) #mean price 
  eth_data['high2mean'] = eth_data['High'] / mean_price 
  eth_data['volume2count'] = eth_data['Volume'] / (eth_data['Count'] + 1)

  return eth_data
```

```python
#dataset with the new features
df = get_features(eth_data)

df
```

#### XGBoost model training and fitting 

We choose the parameters as follows: 
- n_estimators: the number of trees, which will be 100 as the data is high. 

- Max_depth will be 3, which represents the depth of each tree, which represents the maximum number of features used in each tree. I chose a low max depth as by keeping the max_depth low depth will increase incrementally by 1 and it will stop when there is no performance gain in increasing depth. This will simplify the model and hopefully avoid overfitting. 

- Learning rate: will be 0.08, the aim is to slow down the learning rate in order to maintain high performance. 

- Colsample_bytree: the number of columns used by each tree which will be 1,  this will avoid some columns to take too much credit for the prediction, this way we can control variable importance and avoid having the model place too much importance on one variable as the example earlier with random forrest. [29].

```python
import xgboost as xgb
model = xgb.XGBRegressor(objective='reg:squarederror',
    n_estimators=100,
    learning_rate=0.08,
    subsample=0.8,
    colsample_bytree=1,
    max_depth=3)

model.fit(train_features, train_labels) # fitting the model
```

```python
#predicting the model
predictions = model.predict(test_features)
```
